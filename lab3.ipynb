```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df = df[['age','income']]
df.head()
```

```python
df['income'] = df['income'].astype(str).str.strip()
```

```python
grouped = df.groupby('income')['age']
```

```python
summary = grouped.agg(['mean','median','min','max','std'])
summary
```

```python
age_list = grouped.apply(list)
age_list
```
