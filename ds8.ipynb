```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic=sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x='survived', data=titanic)
plt.title("Survival Count (0 = Not Survived, 1 = Survived)")
plt.xlabel("Survival")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Count based on Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(4,3))
plt.hist(titanic['fare'].dropna(), bins=30)
plt.title("Fare Distribution (Histogram)")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title("Fare Distribution across Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.barplot(x='pclass', y='survived', data=titanic)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
sns.pairplot(titanic[['age', 'fare']].dropna())
plt.show()
```

```python
sns.pairplot(titanic[['pclass', 'survived']].dropna())
plt.show()
```
